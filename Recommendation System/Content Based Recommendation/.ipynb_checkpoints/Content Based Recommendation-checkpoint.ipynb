```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movies_df  = pd.read_csv('data/tmdb_5000_movies.csv')
credits_df = pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
movies_df.head()
```

```python
credits_df.head()
```

```python
print(movies_df.shape,credits_df.shape)
# As the shape is same we can merge both dataset
```

```python
# Renaming the movie_id columns in credits_df, to merge both.
credits_df.rename(columns={'movie_id':'id'},inplace=True)
credits_df.head()
```

```python
merged_df = pd.merge(credits_df,movies_df,on='id')
merged_df.head()
```

```python
# Droping unneccesary columns
drop_merged_df =merged_df.drop(columns=['homepage','title_y','title_x','status','production_countries'],axis = 1)
drop_merged_df.head()
```

```python
drop_merged_df.info()
```

```python
drop_merged_df.head(1)['overview']
```

```python
# Tfidf is a concept of NLP
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df = 3,max_features=None,
                     strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',
                     ngram_range=(1,3),
                     stop_words='english')
```

```python
# Filling Nan's with empty string
drop_merged_df['overview']=drop_merged_df['overview'].fillna('')
```

```python
# Fitting the TfidF on 'Overview' Text
tfv_matrix = tfv.fit_transform(drop_merged_df['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
# Compute the sigmoid kernel
kernel = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
kernel[0]
```

```python
# Reverse mapping of indices and movieTitle
indices = pd.Series(drop_merged_df.index,index=drop_merged_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
indices['My Date with Drew']
```

```python
kernel[4802]
```

```python
def give_rec(title, kernel=kernel):
    # Get the index corresponding to original_title
    idx = indices[title]
    
    # Get the pairwise similarity Score
    sig_scores = list(enumerate(kernel[idx]))
    
    # Sort the movies
    sig_scores = sorted(sig_scores,key=lambda x:x[1],reverse=True)
    
    # Scores of 10 most similar movies
    sig_scores=sig_scores[1:11]
    
    # Movie indices
    movie_indices=[i[0] for i in sig_scores]
    
    # top 10 most similar movies
    return drop_merged_df['original_title'].iloc[movie_indices]
```

```python
give_rec('Shanghai Calling')
```
