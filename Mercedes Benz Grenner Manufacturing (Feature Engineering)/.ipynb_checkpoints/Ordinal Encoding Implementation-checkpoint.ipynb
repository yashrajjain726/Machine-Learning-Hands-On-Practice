## Ordinal Encoding Implementation

```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days = x) for x in range(20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
# Extracting the week day name
df['day_of_week'] = df['day'].dt.day_name()
df
```

```python
# Assigning rank to each day of week
day_name_map = {'Monday':1,
                'Tuesday':2,
               'Wednesday':3,
               'Thursday':4,
               'Friday':5,
               'Saturday':6,
               'Sunday':7}
```

```python
df['day_orginal'] = df.day_of_week.map(day_name_map)
df
```
