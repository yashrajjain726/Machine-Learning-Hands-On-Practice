## Missing Value

```python
feature_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
for feature in feature_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
```

### Since there are many missing values, we need to find the relationship between missing values and SalePrice

```python
for feature in feature_with_na:
    data = dataset.copy()
    
    # If the feature has null value,then converting it to 1 and if not than converting it to 0.
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
# We can see from the above diagram'sif NaN values increases,SalePrice is also increasing.
# We can see from the dataset that, Id column is not required
```

### Numerical Variable

```python
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O'] # Where 'O' = Object.
len(numerical_feature)
```

```python
dataset[numerical_feature].head()
```

### Temporary Variable (DateTime Variable)

```python
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
# Let us analyze that there is a relation between YrSold and SalePrice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('SalePrice')
plt.title('House Price Vs Year Sold')
```

```python
for feature in year_feature:
    data = dataset.copy()
    data[feature] = data['YrSold'] - data[feature]
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.show()
```

### Discrete Variable

```python
discrete_feature = [feature for feature in numerical_feature if len(dataset[feature].unique())<25 and 
                    feature not in year_feature + ['Id']]
dataset[discrete_feature].isnull()
```

```python
len(discrete_feature)
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### Continuous Variable

```python
continuous_feature = [feature for feature in numerical_feature if feature not in discrete_feature + year_feature +['Id']]
continuous_feature
```

```python
len(continuous_feature)
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# Converting the graph's into Gaussian Distribution using logrithmic Transformation
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique(): # We are doing this because Log(0) = Undefined, and thus it will ruin the information.
        pass
    else:
        data[feature] = np.log(data[feature]) 
        data['SalePrice'] =np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Checking Outliers

```python
# Converting the graph's into Gaussian Distribution using logrithmic Transformation
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique(): # We are doing this because Log(0) = Undefined, and thus it will ruin the information.
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature) # Remember that boxplot doesn't work for categorical features.
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
       
```
