### We will be performing all the below steps in Feature Engineering
    1. Missing Values
    2. Temporal Variable
    3. Categorical Variable: Remove Rare Labels
    4. Standardise the value of the variables to the same range

### Missing Values

```python
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes == 'O']
for feature in feature_nan:
    print(feature+ " : ",np.round(dataset[feature].isnull().mean(),4))
```

```python
def replace_cat_feature(dataset,feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data
```

```python
dataset = replace_cat_feature(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# Checking Numerical variables containing missing values
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()> 1 
                      and dataset[feature].dtypes !='O' ]
for feature in numerical_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),4))
```

```python
for feature in numerical_with_nan:
    median_value = dataset[feature].median()
    dataset[feature+'nan'] = np.where(dataset[feature].isnull,1,0)
    dataset[feature].fillna(median_value,inplace = True)
```

```python
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# Date-Time Variable
for feature in dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']]:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

### Numerical Variable's

```python
# Found it using Exploratory Data Analysis 
num_features = [['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']]
for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
    
```

```python
dataset.head()
```

### Handling Rare Categorical Feature

```python
# We will remove categorical variables that are present less than 1% of the observations
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset) # getting percentage of that feature
    temp_df = temp[temp > 0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

### Feature Scaling

```python
for feature in categorical_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)

scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice'] ]
len(scaling_feature)
```

```python
scaling_feature
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# Transform the train and test set, and add on the Id and SalePrice variables.
data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop = True),
          pd.DataFrame(scaler.transform(dataset[feature_scale]), columns = feature_scale)],
              axis = 1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv', index = False)
```
