# K- Nearest Neighbour Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Classified Data.csv',index_col = 0)
```

```python
dataset.head()
```

#### As there are many numerical variables in different unit's ,we need to standardize it

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(dataset.drop('TARGET CLASS',axis = 1))
```

```python
scaled_feature = scaler.transform(dataset.drop('TARGET CLASS', axis= 1))
```

```python
scaled_feature
```

```python
dataset_feature = pd.DataFrame(scaled_feature,columns = dataset.columns[:-1])
```

```python
dataset_feature
```

### Pair Plot

```python
import seaborn as sns
sns.pairplot(dataset, hue = 'TARGET CLASS')
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_feature,dataset['TARGET CLASS'], test_size = 0.3)
```

### Using KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
prediction = knn.predict(X_test)
```

```python
prediction
```

### Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

### Choosing a K Value

```python
accuracy_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn, dataset_feature, dataset['TARGET CLASS'], cv = 10)
    accuracy_rate.append(score.mean())
    
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn, dataset_feature, dataset['TARGET CLASS'], cv = 10)
    error_rate.append(1 - score.mean())
    
```

```python
error_rate = []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    predict_i = knn.predict(X_test)
    error_rate.append(np.mean(predict_i != y_test))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,40),error_rate, color = 'blue', linestyle = 'dashed', marker = 'o',
        markerfacecolor = 'red', markersize = 10)
plt.title('Error rate Vs K Value')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

###### Here,you can see that the K > 23, the error rate tends to hover around 0.06 - 0.05 , So When K =  23, the error rate decreases.

```python
knn = KNeighborsClassifier(n_neighbors = 23)
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)
print(confusion_matrix(y_test,prediction))
print(classification_report(y_test,prediction))
```
